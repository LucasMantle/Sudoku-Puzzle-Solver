```python
def sudoku_solver(sudoku):
    found = False
    moves = legal(sudoku) 
    
    if moves == None: 
        return sudoku
    
    #From legal, return the possible numbers aswell so then we dont need to iterate through as many and you know
    #exactly what numbers goes in so no need to call possible 
    
    x_position = moves[0]
    y_position = moves[1]
    pos = moves[2]
    
    for action in pos:
#         if possible(x_position,y_position,action,sudoku):
        sudoku[x_position][y_position] = action
        sudoku = sudoku_solver2(sudoku) 
        #   print(sudoku)
        if np.all(sudoku != 0):
            found = True
            return sudoku
        sudoku[x_position][y_position] = 0
    return np.negative(np.ones((9, 9)))  

def sudoku_solver2(sudoku):
    found = False
    moves = legal(sudoku) 
    if moves == None: 
        return sudoku
    x_position = moves[0]
    y_position = moves[1]
    pos = moves[2]
    
    for action in pos:
        #if possible(x_position,y_position,action,sudoku):
        sudoku[x_position][y_position] = action
        sudoku = sudoku_solver2(sudoku) 
        #   print(sudoku)
        if np.all(sudoku != 0):
            found = True
            return sudoku
        sudoku[x_position][y_position] = 0
    return sudoku  
    
def possible (x,y,n, sudoku):
    grid = sudoku
    for i in range(0,9):
        if grid[x][i] == n:
            #print("x")
            return False
            
    for i in range (0,9):
        if grid[i][y] == n:
            #print("z")
            return False
    x0 = (x//3)*3
    y0 = (y//3)*3
    for i in range(0,3):
        for j in range(0,3):
            if grid[x0+i][y0+j] == n:
                #print("y")
                return False
    return True

    
def legal(sudoku):
    legal = np.zeros(81)
    lists = []
    for x in range(9):
        for y in range(9):
            store = []
            if sudoku[x][y] == 0:
                for n in range(1,10):
                    #print(sudoku)
                    if possible(x,y,n,sudoku):
                        legal[(x*9)+(y)] = legal[(x*9)+(y)] + 1
                        store.append(n)
            
            lists.append(store)
    
    
    if np.all(legal == 0):
        return None
    else: 
        
        l = legal
        m = np.min(l[np.nonzero(l)])
        itemindex = np.where(l==m)
        z = itemindex[0][0]
        pos = lists[z]
        x_position = int(z/9)
        y_position = int(z%9)
        return (x_position,y_position,pos)
```
